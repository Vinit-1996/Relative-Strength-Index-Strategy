RSI

The Relative Strength Index will allow you to understand the strength of a trend. The RSI Value will vary from 0 to 100.



*   If the RSI is around the 70 level, it is assumed that the instrument is overbrought. This is a situation where prices have risen sharply, and the movement is likely to weaken.
*   If the RSI is around the 30 level, this is a technical signal that the instrument may be oversold. In this case, it is the situation where prices have fallen sharply and the movement may now lose strength.
*  If the indicator ocisllate horizontally around 50 level, it means that the market is lacking a tren. The 50 level is the middle line that seperates the bullish and bearish territories of the indicator.

It should be taken into account that the RSI is an indicator of the oscillator family, so it will work a little less well in markets with less volatility such as the stock market.




```python
#Import the data

f=yf.download(["GOOG"])
```

```python
f.head()
```

```python
f["RSI"]=ta.momentum.RSIIndicator(f["Adj Close"],window=14).rsi()
```

```python
f.tail()
```

```python
#Plot the RSI with the overbuy and oversell threshold.

plt.figure(figsize=(15,8))

plt.plot(f['RSI'].loc["2010"])

plt.axhline(70,color="#57CE95")


plt.axhline(30,color="#CE5757")

plt.title("RSI with threholds")

plt.legend(['RSI','Overbuy threshold','Oversell Threshold'])

plt.show()
```

```python
#Plot the RSI with the long buy and short sell zone
plt.figure(figsize=(15,8))

plt.plot(f['RSI'].loc["2010"])

plt.fill_between(f['RSI'].loc["2010"].index,55,70,55<70,color='#57CE95',alpha=0.5)

plt.fill_between(f['RSI'].loc["2010"].index,45,30,30<45,color='#CE5757',alpha=0.5)

plt.title("RSI with zone")

plt.legend(['RSI','Long buy zone','Short sell zone'])

plt.show()
```

```python
# We define when the strategy must take a long buy or a short sell

# we need define the Open long and Close long signal

# We put threshold

overbuy=70
neutral_buy=55

#put nan values for the signal long columns
f['signal_long']=np.nan
f['yesterday_rsi']=f['RSI'].shift(1)

#We need to define the Open long signal (RSI yesterday<55 and RSI today>55)

f.loc[(f['RSI']>neutral_buy) & (f['yesterday_rsi']>neutral_buy),'signal_long']=1

#We need to define the Close Long Signal (RSI Yesterday >55 and RSI Today <55) False signal

f.loc[(f['RSI']<neutral_buy) & (f['yesterday_rsi']>neutral_buy),'signal_long']=0

#We need to define the Close Long Signal (RSI Yesterday >70 and RSI Today <70) Over Buy Signal

f.loc[(f['RSI']<overbuy) & (f['yesterday_rsi']>overbuy),'signal_long']=0
```

```python
#We plot all the signals to be sure that they are correct

#Select all signals in a index list to plot only this point

idxopen=f.loc[f["signal_long"]==1].loc["2010"].index

idxclose=f.loc[f["signal_long"]==0].loc["2010"].index

plt.figure(figsize=(15,8))

plt.scatter(f.loc[idxopen]["RSI"].index, f.loc[idxopen]["RSI"].loc["2010"],color='#57CE95',marker='^')

plt.scatter(f.loc[idxclose]["RSI"].index, f.loc[idxclose]["RSI"].loc["2010"],color='#669FEE',marker='o')

plt.plot(f["RSI"].loc["2010"].index, f['RSI'].loc['2010'],alpha=0.35)

plt.show()
```

```python
# We define when the strategy must take a long buy or a short sell

# we need define the Open Short and Close short signal

# We put threshold

oversell=30
neutral_sell=45

#put nan values for the signal short columns
f['signal_short']=np.nan
f['yesterday_rsi']=f['RSI'].shift(1)

#We need to define the Open short signal (RSI yesterday>45 and RSI today<45)

f.loc[(f['RSI']<neutral_sell) & (f['yesterday_rsi']>neutral_sell),'signal_short']=-1

#We need to define the Close Short Signal (RSI Yesterday <45 and RSI Today >45) False signal

f.loc[(f['RSI']>neutral_sell) & (f['yesterday_rsi']<neutral_sell),'signal_short']=0

#We need to define the Close Long Signal (RSI Yesterday >70 and RSI Today <70) Over Buy Signal

f.loc[(f['RSI']>oversell) & (f['yesterday_rsi']<oversell),'signal_short']=0
```

```python
#We plot all the signals to be sure that they are correct

#Select all signals in a index list to plot only this point

idx_open=f.loc[f["signal_short"]==-1].loc["2010"].index

idx_close=f.loc[f["signal_short"]==0].loc["2010"].index

plt.figure(figsize=(15,8))

plt.scatter(f.loc[idx_open]["RSI"].index, f.loc[idx_open]["RSI"].loc["2010"],color='#57CE95',marker='^')

plt.scatter(f.loc[idx_close]["RSI"].index, f.loc[idx_close]["RSI"].loc["2010"],color='#669FEE',marker='o')

plt.plot(f["RSI"].loc["2010"].index, f['RSI'].loc['2010'],alpha=0.35)

plt.show()
```
